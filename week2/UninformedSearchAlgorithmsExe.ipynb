# Breadth-First Search (BFS)

```python
Image("img/BFS.png")
```

```python
def bfs(initialStateName, goalStateName):
    frontier = deque()
    frontier.append(initialStateName)
    explored = deque()
    while frontier:
        stateName = frontier.popleft()
        explored.append(stateName)
        if stateName == goalStateName:
            print("#### Answer:\n")
            print (*explored, sep =', ') 
            return True
    
        for neighborName in sorted(G[stateName].keys()):
            if (neighborName not in frontier) and (neighborName not in explored):
                frontier.append(neighborName)
        
        print("Current Queue:\n",frontier)
        print()
        print("Current Explored:\n",explored)
        print()
    return False
```

```python
bfs("San Francisco", "Seattle")
```

# Depth-First Search (DFS)

```python
Image("img/DFS.png")
```

```python
def dfs(initialStateName, goalStateName):
    frontier = deque()
    frontier.append(initialStateName)
    explored = deque()
    while frontier:
        stateName = frontier.pop()
        explored.append(stateName)
        if stateName == goalStateName:
            print("#### Answer:\n")
            print (*explored, sep =', ') 
            return True
    
        for neighborName in sorted(G[stateName].keys(),reverse=True):
            if (neighborName not in frontier) and (neighborName not in explored):
                frontier.append(neighborName)
        
        print("Current Queue:\n",frontier)
        print()
        print("Current Explored:\n",explored)
        print()
    return False
```

```python
dfs("Chicago", "Vancouver")
```

# Uniformed Cost Search (UCS)

```python
Image("img/UCS.png")
```

```python
def ucs(initialStateName, goalStateName):
    frontier = list()
    frontier.append((initialStateName,0))
    explored = deque()    
    paths = list()
    while frontier:
        stateName, cost = frontier.pop(0)
        explored.append((stateName, cost))
        if stateName == goalStateName:
            answerNames = [n for (n,c) in explored]
            print("#### Answer:\n")
            print (*answerNames, sep =', ') 
            return True
        neighborsNodes = G[stateName]
        for neighborName in sorted(G[stateName].keys()):
            neighborWeight =  neighborsNodes[neighborName]['weight'] + cost
            if (neighborName not in str(frontier)) and (neighborName not in str(explored)):
                frontier.append((neighborName,neighborWeight))
            elif (neighborName in str(frontier)):
                for (storedName, storedWeight) in frontier:
                    if (storedName == neighborName) and (neighborWeight < storedWeight):
                        frontier.remove((storedName, storedWeight))
                        frontier.append((neighborName,neighborWeight))
                            
        frontier = sorted(frontier, key=lambda tup: tup[1])
        print("Current Queue:\n",frontier)
        print()
        print("Current Explored:\n",explored)
        print()
    return False
```

```python
ucs('Montreal', 'Duluth')
```
